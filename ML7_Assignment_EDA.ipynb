## Q1. What are the key features of the wine quality data set? Discuss the importance of each feature in
# predicting the quality of wine.
Ans.
1. Fixed Acidity: Fixed acidity represents the total amount of acids in the wine, which contribute to the wine's taste and 
stability. It's an important indicator of the wine's overall acidity level, which affects its tartness and crispness.

2. Volatile Acidity: Volatile acidity measures the presence of volatile acids, primarily acetic acid, in the wine. Too high
a level can result in a vinegary taste and spoil the wine's quality.

3. Citric Acid: Citric acid is a natural acid found in many fruits and is often added to wines to improve freshness and flavor
balance. It can contribute to the wine's acidity and enhance its taste profile.

4. Residual Sugar: Residual sugar refers to the amount of sugar remaining in the wine after fermentation. It affects the wine's
sweetness and can influence the perceived balance between acidity and sweetness.

5. Chlorides: Chloride concentration can impact the taste and perception of saltiness in the wine. Proper levels contribute to 
the wine's overall taste and structure.

6. Free Sulfur Dioxide: Free sulfur dioxide is added to wine as a preservative and antioxidant. It helps prevent unwanted 
microbial growth and oxidation.

7. Total Sulfur Dioxide: Total sulfur dioxide includes both free and bound forms of sulfur dioxide. It's important for the 
wine's stability, quality, and longevity.

8. Density: Density is an indicator of the wine's mass per unit volume. It's influenced by the concentration of various components,
such as alcohol and sugar, and can provide insights into the wine's composition.

9. pH: pH measures the acidity or basicity of the wine. It affects the wine's taste, color, and stability. Lower pH values indicate 
higher acidity.

10. Sulphates: Sulphates (SO4) are often added to wines as a preservative and stabilizer. They can contribute to the wine's flavor,
mouthfeel, and overall quality.

11. Alcohol: Alcohol content significantly influences the wine's body, flavor, and mouthfeel. It contributes to the wine's perceived
warmth and complexity.

These features are important in predicting wine quality because they collectively provide information about the chemical composition of 
the wine.

```python
# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
# features. What is the minimum number of principal components required to explain 90% of the variance in
# the data?

import pandas as pd
df = pd.read_csv('winequality-red.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[::,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.30,random_state=20)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# Standard scaler 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
# performing PCA
from sklearn.decomposition import PCA
pca = PCA(0.90)
pca.fit(X_train_scaled)
print(pca.explained_variance_ratio_)
```

```python
# Transforming X_train and X_test
x_train_pca = pca.transform(X_train_scaled)
x_test_pca = pca.transform(X_test_scaled)
```

```python
x_train_pca
```

```python
x_test_pca
```

```python
print(f'Minimum Number of components required to explain 90% variance is {x_train_pca.shape[1]}')
```
